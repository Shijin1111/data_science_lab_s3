```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x,y,label = "y=sin(x)",color="green")
plt.xlabel("x", fontsize=12, labelpad=10)  
plt.ylabel("y", fontsize=12, labelpad=10)  
plt.title("y=sin(x)")
plt.legend()
plt.grid(True)
plt.show() #works without plt.show()
```
